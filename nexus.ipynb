```python
import os

TRAIN_PATH = '/content/drive/MyDrive/Train_Data'
keywords = ['BenchPress', 'BodyWeightSquats', 'CleanAndJerk', 'HandstandPushups',
            'JumpingJack', 'Lunges', 'PullUps', 'PushUps', 'WallPushups']

def clean_filenames(path):
    print(f"🧹 Cleaning files in {path}...")
    files = [f for f in os.listdir(path) if f.endswith('.npy')]
    count = 0

    for old_name in files:
        new_name = old_name
        # 1. Remove the 'v_' ghost
        if '_v_' in new_name:
            new_name = new_name.replace('_v_', '_')

        # 2. Remove redundant double-names (e.g., PullUps_PullUps -> PullUps)
        for kw in keywords:
            double_kw = f"{kw}_{kw}"
            if double_kw in new_name:
                new_name = new_name.replace(double_kw, kw)

        # 3. Rename the actual file in Drive
        if new_name != old_name:
            os.rename(os.path.join(path, old_name), os.path.join(path, new_name))
            count += 1

    print(f"✅ Successfully cleaned {count} filenames!")

clean_filenames(TRAIN_PATH)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import numpy as np
import os

ACTION_MAP = {
    'BenchPress': 0, 'BodyWeightSquats': 1, 'CleanAndJerk': 2,
    'HandstandPushups': 3, 'JumpingJack': 4, 'Lunges': 5,
    'PullUps': 6, 'PushUps': 7, 'WallPushups': 8
}

class NexusDataset(Dataset):
    def __init__(self, path):
        self.path = path
        # Filter files: only those starting with our keys and ending in .npy
        self.files = [f for f in os.listdir(path) if f.split('_')[0] in ACTION_MAP and f.endswith('.npy')]

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fname = self.files[idx]
        data = np.load(os.path.join(self.path, fname)) # Shape: (100, 17, 2)

        # 1. FIX DIMENSION MISS-MATCH: (100, 17, 2) -> (2, 100, 17)
        data = data.transpose(2, 0, 1)

        # 2. LABEL MAPPING
        label_name = fname.split('_')[0]
        label = ACTION_MAP[label_name]

        # 3. HIP-CENTERED NORMALIZATION (Crucial for 80%+ Accuracy)
        # We subtract the first joint (usually hip) from all joints
        data[0] -= data[0, :, 0:1]
        data[1] -= data[1, :, 0:1]

        # 4. GLOBAL SCALING
        max_val = np.max(np.abs(data)) + 1e-6
        data /= max_val

        # 5. GAUSSIAN JITTER (Only for Train_Data)
        # This prevents the model from "cheating" and memorizing exact positions
        if 'Train_Data' in self.path:
            noise = np.random.normal(0, 0.005, data.shape)
            data += noise

        return torch.from_numpy(data).float(), label

# --- INITIALIZE LOADERS ---

# Loader 1: Training (Noisy & Shuffled)
train_ds = NexusDataset('/content/drive/MyDrive/Train_Data')
train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)

# Loader 2: Testing (Clean & Ordered)
# Use this for your 79% test accuracy checks
test_ds = NexusDataset('/content/drive/MyDrive/Test_Data')
test_loader = DataLoader(test_ds, batch_size=1, shuffle=False)

print(f"✅ Data Ready! Training samples: {len(train_ds)} | Testing samples: {len(test_ds)}")
```

```python
import torch.nn as nn

class STGCN_Lite(nn.Module):
    def __init__(self, num_class=9):
        super(STGCN_Lite, self).__init__()
        # Input: [Batch, 2, 100, 17]
        self.bn = nn.BatchNorm1d(2 * 17)

        # LAYER 1: Focus on Time (The movement rhythm)
        self.conv1 = nn.Sequential(
            # Using (9, 1) kernel to see 9 frames at once instead of 3
            nn.Conv2d(2, 64, kernel_size=(9, 1), padding=(4, 0)),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d((2, 1)) # Down to 50 frames
        )

        # LAYER 2: Focus on Joints (The skeleton shape)
        self.conv2 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=(1, 3), padding=(0, 1)),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d(1)
        )

        self.fc = nn.Linear(128, num_class)

    def forward(self, x):
        N, C, T, V = x.size()

        # Normalization Step
        x = x.permute(0, 1, 3, 2).contiguous().view(N, C*V, T)
        x = self.bn(x)
        x = x.view(N, C, V, T).permute(0, 1, 3, 2).contiguous() # [N, 2, 100, 17]

        x = self.conv1(x)
        x = self.conv2(x)
        x = x.view(N, -1)
        return self.fc(x)

print("✅ Optimized STGCN_Lite Defined.")
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ASSIGN THE MODEL
model = STGCN_Lite(num_class=9).to(device)

# TEST THE DIMENSIONS
sample_input, _ = train_ds[0]
sample_input = sample_input.unsqueeze(0).to(device) # Add batch dimension

model.eval()
with torch.no_grad():
    output = model(sample_input)

print(f"✅ Handshake Successful! Output shape: {output.shape}")
# Should print [1, 9]
```

```python
import torch.optim as optim

# 1. Setup with a SMARTER Learning Rate
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)
best_acc = 0.0

print("🚀 Starting Final Training for nexus...")

for epoch in range(1, 101):
    model.train()
    running_loss, correct, total = 0.0, 0, 0

    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

    # Calculate Accuracy
    epoch_acc = 100. * correct / total

    if epoch % 5 == 0 or epoch == 1:
        print(f"Epoch {epoch:03d} | Loss: {running_loss/len(train_loader):.4f} | Acc: {epoch_acc:.2f}%")

    # SAVE THE BEST MODEL AUTOMATICALLY
    if epoch_acc > best_acc:
        best_acc = epoch_acc
        torch.save(model.state_dict(), '/content/drive/MyDrive/nexus_stgcn_final.pth')
        if epoch_acc > 70:
            print(f"⭐ New High Score: {best_acc:.2f}% - Model Saved!")

print(f"🏆 Training Finished! Best Accuracy: {best_acc:.2f}%")
```

```python
# 1. Load the "Final" Weights from your Drive
# Make sure the filename matches what you saved in the training loop!
model_path = '/content/drive/MyDrive/nexus_stgcn_final.pth'

if os.path.exists(model_path):
    model.load_state_dict(torch.load(model_path))
    print("🎯 Successfully loaded the Final Model weights.")
else:
    print("❌ ERROR: Model file not found. Check your Drive path!")

model.to(device)
model.eval()

# 2. Setup Test Dataset (Ensure no jitter/noise is applied here)
test_ds = NexusDataset('/content/drive/MyDrive/Test_Data')
test_loader = DataLoader(test_ds, batch_size=1, shuffle=False)

test_correct = 0
total_samples = len(test_ds)

print(f"🧪 Testing on {total_samples} unseen videos...")

with torch.no_grad():
    for inputs, labels in test_loader:
        # Move data to the same device as the model (GPU)
        inputs = inputs.to(device)
        labels = labels.to(device)

        # Forward pass
        outputs = model(inputs)

        # Get the index of the highest score
        _, predicted = outputs.max(1)

        # Compare
        if predicted.item() == labels.item():
            test_correct += 1

# 3. Final Calculation
final_test_acc = 100. * test_correct / total_samples
print("-" * 30)
print(f"✅ Final Test Accuracy: {final_test_acc:.2f}%")
print(f"📊 Correct: {test_correct} | Incorrect: {total_samples - test_correct}")
```

```python
from collections import defaultdict

class_correct = defaultdict(int)
class_total = defaultdict(int)
inv_map = {v: k for k, v in ACTION_MAP.items()}

model.eval()
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = outputs.max(1)

        label_name = inv_map[labels.item()]
        class_total[label_name] += 1
        if predicted.item() == labels.item():
            class_correct[label_name] += 1

print("📊 --- PER-CLASS ACCURACY REPORT ---")
for label in ACTION_MAP.keys():
    acc = 100 * class_correct[label] / class_total[label] if class_total[label] > 0 else 0
    print(f"{label:18s}: {acc:.2f}% ({class_correct[label]}/{class_total[label]})")
```

```python
import random
import torch.nn.functional as F
import os
import numpy as np

def live_demo(test_path):
    # 1. Load the NEW "Final" Model from your Drive
    # We use the 'final' file you are saving in your 100-epoch loop
    model_path = '/content/drive/MyDrive/nexus_stgcn_final.pth'

    if not os.path.exists(model_path):
        print(f"⚠️ Model file not found at {model_path}. Check your Drive!")
        return

    model.load_state_dict(torch.load(model_path))
    model.eval()

    # 2. Pick a random test file
    test_files = [f for f in os.listdir(test_path) if f.endswith('.npy')]
    if not test_files:
        print("Empty folder! Check your Drive path.")
        return
    random_file = random.choice(test_files)

    # 3. Load and Transform (Must match training exactly)
    data = np.load(os.path.join(test_path, random_file))
    raw_label = random_file.split('_')[0]

    # Dimension Flip: (100, 17, 2) -> (2, 100, 17)
    input_data = data.transpose(2, 0, 1)

    # Normalization (Hip-centering)
    input_data[0] -= input_data[0, :, 0:1]
    input_data[1] -= input_data[1, :, 0:1]
    input_data /= (np.max(np.abs(input_data)) + 1e-6)

    # Convert to Tensor
    input_tensor = torch.from_numpy(input_data).float().unsqueeze(0).to(device)

    # 4. Predict with Confidence
    with torch.no_grad():
        output = model(input_tensor)
        probabilities = F.softmax(output, dim=1)
        conf, predicted = probabilities.max(1)

    # 5. Map back to Name
    inv_map = {v: k for k, v in ACTION_MAP.items()}
    prediction = inv_map[predicted.item()]

    # 6. The Big Reveal
    print("-" * 30)
    print(f"🎬 Testing Video: {random_file}")
    print(f"🎯 Actual Exercise: {raw_label}")
    print(f"🤖 AI Prediction:   {prediction} ({conf.item()*100:.1f}% Confidence)")

    if prediction == raw_label:
        print("✅ SUCCESS: Match Found!")
    else:
        print(f"❌ MISMATCH: AI was {conf.item()*100:.1f}% sure, but missed.")
    print("-" * 30)

# Run the demo against your Test folder
live_demo('/content/drive/MyDrive/Test_Data')
```
